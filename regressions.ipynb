```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy as stats
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('housing.xlsx')
data
```

```python
X=data['House Size (sq.ft.)']
```

```python
Y=data['House Price']
```

```python
X1=sm.add_constant(X)
```

```python
reg=sm.OLS(Y,X1).fit()
```

```python
reg.summary()
```
